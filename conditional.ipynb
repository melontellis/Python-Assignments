```python
# Check given no is positive, zero, negative

if else ladder/continous if else syntax

if condition1:
    statements of condition1
else:
    if condition2:
        statements of condition2
    else:
        if condition3:
            statements of condition3
            ....
            else:
            default statements
```

```python
if condition1:
    statements of condition1
elif condition2:
    statements of condition2
elif condition3:
    statements of condition3
    ....
else:
    default statements
```

```python
n1 = int(input("enter number"))
if (n1>0):
    print(n1, "is positive")
elif (n1==0):
    print(n1, "is zero")
else:
    print(n1,"is negative")
```

```python
# based on percentage print the result
# 85% - 100% =>  Distinction
# 60 - 84 = 1st class
# 50 - 59 = second class
# >35% - 49 = juat pass
# <=35 => fail

n = float(input("Enter marks"))
if(n<0 or n>100):
    print("Invalid Input")
elif (n>=85):
    print("Distinction")
elif (n>=60):
    print("First Class")
elif (n>=50):
    print("Second Class")
elif (n>=35):
    print("Just Pass")
else:
    print("Fail")
```

```python
# Check the given input that they form which type of triangle
# Equilateral
# Isosceles
# Scalene

a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))

if a == b == c:
        print("Equilateral Triangle")
elif a == b or b == c or a == c:
        print("Isosceles Triangle")
else:
        print("Scalene Triangle")
```

```python
nested if else syntax
if inside another if called as nested if

if condition1: #outer if
    if condition2:  #inner if
    statements condition2
    else:
    statements of inner else
else:
statements of outer else
```

```python
n = float(input("Enter marks"))
if(n<0 or n>100):
    print("Invalid Input")
elif (n>=85):
    if(n>=95):
        print("Super Distinction")
    else:
        print("Distinction")
elif (n>=60):
    print("First Class")
elif (n>=50):
    print("Second Class")
elif (n>=35):
    print("Just Pass")
else:
    print("Fail")
```
