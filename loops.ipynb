```python
# Write a prg to check the given no is prime or not

n = int(input("Enter a number: "))
count = 0

for i in range(2, n):
    if n % i == 0:
        count += 1   

if count == 0:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
# write prg to calculate sum of digits
# 145 = 1+4+5 = 10

n = int(input("Enter a number: "))
sum = 0

while n > 0:
    digit = n % 10        
    sum += digit      
    n = n // 10   

print("Sum of digits =", sum)
```
