```python
# Write a prg to check the given no is prime or not

n = int(input("Enter a number: "))
count = 0

for i in range(2, n):
    if n % i == 0:
        count += 1   

if count == 0:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
# write prg to calculate sum of digits
# 145 = 1+4+5 = 10

n = int(input("Enter a number: "))
sum = 0

while n > 0:
    digit = n % 10        
    sum += digit      
    n = n // 10   

print("Sum of digits =", sum)
```

```python
# write a program to check the given number is palindrome or not
# 121 => 121 its palindrome
# 123 => 321 its not palindrome

n = int(input("enter a number:"))
temp = n
print("origianl=", temp)
rev = 0

while(n > 0):
    rem = n % 10
    rev = (rev * 10) + rem
    n = n // 10

print("reverse=", rev)

if(temp == rev):
    print("its a palindrome")
else:
    print("not a palindrome")
```

```python
# write a program to find factorial of given number
# 7*6*5*4*3*2*1

n = int(input("Enter a number: "))
fact = 1

while n > 1:
    fact = fact * n
    n = n - 1

print("Factorial =", fact)
```
