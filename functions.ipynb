```python
def prime():
    sv = int(input("Enter a start value:"))
    sp = int(input("Enter a stop value:"))
    for i in range(sv, sp+1, 1):
        for j in range (2,i,1):
            if(i%j==0):
                break
        else:
            print()
```

```python
def prime():
    pl = []
    for i in range(sv, sp+1, 1):
        for j in range (2,i,1):
            if(i%j==0):
                break
        else:
            pl.append(i)
            return pl
```

```python
lambda function
a lambda function is small, anonymous(nameless) function in python

*it defined using keyword lambda instead of def
*it can take any number of arguements but must contain only one expression
*expression is automatically returned (no need to use return)

syntax 
lambda arguements:expression
```

```python
sq = lambda x: x*x
sq(2)
```

```python
add = lambda a,b : a+b
add(3,9)
```

```python
why to use lambda?
*useful for short, simple operation
• saves code(one-liner)
• often used with map(),filter(),reduce()
• or as inline function

When NOT use lambda?
• for complex logic(use def insted)
• when readbility is more important than shortness.
```

```python
Create function to calculate factorial of given number using without input with return
```

```python
def factorial():
    n = int(input("Enter a number: "))
    fact = 1

    for i in range(n, 0, -1):
        fact = fact * i

    return fact
```

```python
factorial()
```
