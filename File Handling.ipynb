```python
What is file handling?
It will allow us to create, read, write, update and delete files.
It is used to store data permanently/temp.
The built-in function is open() to work with files
Common modes:
1. "r" - read
2. "w" - write
3. "a" - append
4. "b" - binary
5. "t" - text
6. "rb" - read binary
7. "wt" - write text
File path:
relative path: file name
absolute path: directories(complete loc) + file name
```
