```python
Here is the text extracted exactly from the image:
]: on insurance data
1) each region wise total expenses
2) gender wise avg bmi and expenses
3) each region, each children and somker class wise total expenses
use: group by and pivot table
```

```python
inu = pd.read_csv("D:\Anaconda\Maam Data\insurance.csv")
inu
```

```python
inu.groupby(by ='region')
```

```python
inu.groupby(by ='region')['expenses'].sum()
```

```python
inu['expenses'].sum()
```

```python
pd.pivot_table(data = inu, values ='expenses', index ='region' , aggfunc ='sum')
```

```python
inu.groupby(by ='region')['expenses'].sum().sort_values()
```

```python
inu.groupby(by ='region')['expenses'].sum().sort_values(ascending = False)
```

```python
inu.groupby(by ='region')['expenses'].sum().sort_values()
```

```python
# to sort region 
```

```python
inu.groupby(by ='region',sort = 'asc')['expenses'].sum()
```

```python
# gender wise avg bmi and expenses
inu.groupby(by = 'sex')[['bmi','expenses']].mean().round(2)
```

```python
pd.pivot_table(data = inu, values = ['bmi','expenses'],index = 'sex', aggfunc= 'mean').round(2)
```

```python
# each region, each children and somker class wise total avg expenses
inu.groupby(by = ['region', 'children', 'smoker'])['expenses'].mean().round(2)
```

```python
pd.pivot_table(data = inu, values = 'expenses', index = 'region', columns = ['smoker', 'children'], aggfunc = 'mean').round(2)
```
