```python
s1 = "Besant$184#209@GmaIL.CoM"

count of
upper char
lower char
numbers
special char
```

```python
s1 = "Besant$184#209@GmaIL.CoM"
uc = 0
lc = 0
nc = 0
sc = 0
for i in s1:
    if i.isupper():
        uc+=1
    elif i.islower():
        lc+=1
    elif i.isdigit():
        nc+=1
    elif i!=" ":
        sc+=1
print(f"Upper count:{uc}\nLower count:{lc}\nNumber count:{nc}\nSpecial Char count:{sc}")
```
