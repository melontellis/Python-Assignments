```python
What is data visualization?
it is the process of representing data in a grafical or visual format such as charts, graphs or maps so that people can easily understand patterns,
trends and insights.
    
Why visualization is important
 *makes data easy to understand
 *helps in quick decision making
 *isentifies patterns and trends
 *detects outliers or errors
 *improves communication of insights
```

```python
Univariate analysis:
It means analyzing only one variable at a time.
to check the quality of the data.

    Purpose:
 *understand distribution
 *find patterns
 *detect outliers
ex: Histogram, Box plot, KDE plot, Violin plot
```

```python
Bivariate analysis:
It means analyzing the relationship between two variables.

    Purpose:
 * understand relationship
 * identify correlation
 * compare variables
ex: scatter plot, bar plot, line plot
ex: analyze advertising spend Vs sales
Does increasing ads increase sales?
```

```python
Multivariate analysis:
it means analyzing relationship between more than two variables.

    Purpose:
 * undestand complex relationships
 * identify patterns across multiple variables
 * perform deep data analysis(EDA)
EDA: exploratory data analysis
ex: Pair plot, heatmap, facet grid, bubble plot
```

```python
Matplotlib and Seaborn
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
Histogrsm:
* Shows distribution of data
* Helps to detect skewness amd spread
```

```python
plt.figure()
plt.hist(df['total_bill'])
plt.title("histogram")
plt.xlabel("Total bill")
plt.ylabel("Frequencty")
plt.show()
```

```python
df['total_bill'].max()
```

```python
KDE plot: (kernal density estimation plot)
*smooth version of histogram
*better for understanding data density
```

```python
plt.figure()
sns.kdeplot(df['total_bill'], fill=True)
plt.title("KDE Plot")
plt.xlabel("Total Bill")
plt.ylabel("frequnecy")
plt.show()
```

```python
Box plot:
*identifies outliers
*shows median, quartiles, spread
```

```python
plt.figure()
sns.boxplot(x=df['total_bill'])
plt.title("Box Plot")
plt.xlabel("Total Bill")
plt.show()
```

```python
Violin plot:
*combines box plot + distribution
*shows data shape clearly
```

```python
plt.figure()
sns.violinplot(x=df['total_bill'])
plt.title("Violin plot")
plt.show()
```

```python
Bar plot:
*compares values across categories
*easy to understand differences
```

```python
plt.figure()
sns.barplot(x="day", y="total_bill", data=df, orient='v')
plt.title("Bar plot")
plt.show()
```

```python
Pie chart:
*shows proportion/percentage
*good for quick insights
note: use only for small num of categories
```

```python
plt.figure()
df['day'].value_counts().plot.pie(autopct='%1.2f%%')
plt.title("pie chart")
plt.show()
```

```python
line plot:
*shows trends over time
*helps to identify increases, decrease, seasonality
ex: sales growth month-wise
```

```python
plt.plot(df['total_bill'])
plt.title("Line Plot")
plt.show()
```

```python
Count Plot
*shows frequency of categories
*useful for categorical analysis
ex: number of customers per city
```

```python
sns.countplot(x='day', data=df)
plt.title("count Plot")
plt.show()
```

```python
scatter plot:
*shows relationship between two numerical variables
*helps detect correlation and patterns
ex: Advertising Vs Sales
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("scatter plot")
plt.show()
```

```python
bubble sort:
*add third variable(size)
*shows more information in one chart
ex: sales Vs Profit with market size
```

```python
plt.scatter(df['total_bill'], df['tip'], s=df['size']*20)
plt.title("Bubble plot")
plt.show()
```

```python
Pair Plot:
 shows relationship between all variables
 useful in EDA(exploratory data analysis)
ex: feature relationships in ML
```

```python
sns.pairplot(df)
plt.show()
```

```python
Heatmap:
*shows correlationship between variable
*helps in feature selection
ex: identity highly corelated features
```

```python
corr = df.corr(numeric_only=True)
corr.round(2)
```

```python
sns.heatmap(corr,annot = True)
plt.show()
```
